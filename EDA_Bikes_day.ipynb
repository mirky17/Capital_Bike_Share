```python
day['date']= pd.to_datetime(day['date'],format='%Y-%m-%d')
day.head()
```

```python
# renaming values to make them look appropriate.

#Holidays:
day.loc[day['holiday'] == 0, 'holiday'] = 'No'
day.loc[day['holiday'] == 1, 'holiday'] = 'Yes'
```

```python
#Season
day.loc[day['season'] == 1, 'season'] = 'Winter'
day.loc[day['season'] == 2, 'season'] = 'Spring'
day.loc[day['season'] == 3, 'season'] = 'Summer'
day.loc[day['season'] == 4, 'season'] = 'Fall'
```

```python
#workingday
day.loc[day['workingday'] == 1, 'workingday'] = 'Yes'
day.loc[day['workingday'] == 0, 'workingday'] = 'No'
day.head()
```

```python
#year
day.loc[day['year'] == 1, 'year'] = '2012'
day.loc[day['year'] == 0, 'year'] = '2011'
day.head()
```

```python
m=sns.FacetGrid(day, hue='year', palette= 'icefire', size=6,aspect=2)
m=m.map(plt.hist,'count',alpha=0.6, edgecolor='w')
plt.legend()
title = 'Total Of Rental Bikes over Years'
plt.title(title)
```

```python
plt.figure(figsize=(11,5))
sns.barplot('year','count',hue='season', data=day,palette='coolwarm', ci=None)
plt.legend(loc='upper right',bbox_to_anchor=(1.2,0.5))
plt.xlabel('Year')
plt.ylabel('Total number of bikes rented')
plt.title('Number of bikes rented per season')
```

```python
plt.figure(figsize=(11,5))
sns.barplot('year','count',hue='holiday', data=day,palette='coolwarm', ci=None)
plt.legend(loc='upper right',bbox_to_anchor=(1.2,0.5))
plt.xlabel('Year')
plt.ylabel('Total number of bikes rented')
plt.title('Number of bikes rented holiday-Not holiday')
```
